```python
import pandas as pd
df=pd.read_csv('BlackFriday.csv')
df.head()
```

```python
df.groupby('Age').size().plot(kind= 'bar' , figsize=(12,6), title='Purchase Distribution By Age')
```

```python
lst=[]
for i in df['Age'].unique():
    lst.append([i,df[df['Age']==i]['Product_ID'].nunique()])
data=pd.DataFrame(lst,columns=['Age','Product'])
data.plot.bar(x='Age',figsize=(8,6))
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind= 'bar' , figsize=(12,6), title='Amount Spend By Age')
```

```python
df.groupby('Age').mean(numeric_only=True)['Purchase'].plot(kind= 'bar' , figsize=(12,6), title='Average Amount Spend By Age')
```

```python
df.groupby('Age').mean(numeric_only=True)['Purchase']
```

```python
df.groupby('Age').mean(numeric_only=True)['Purchase'].plot(kind='pie', autopct='%.2f')
```
