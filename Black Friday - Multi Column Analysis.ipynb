```python
import pandas as pd
import seaborn as sns
df=pd.read_csv('BlackFriday.csv')
df.head()
```

```python
df.groupby('Age').size().plot(kind= 'bar' , figsize=(12,6), title='Purchase Distribution By Age')
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.countplot(data=df , x='Age', hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.countplot(data=df , x='Gender', hue='Age')
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.countplot(data=df , x='Gender', hue='Marital_Status')
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.countplot(data=df , x='Marital_Status', hue='Gender')
```

```python
sns.countplot(x=df['City_Category'])
```

```python
df.groupby('City_Category').size().plot(kind='pie', autopct='%.2f')
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.countplot(data=df , x='City_Category', hue='Age')
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.countplot(data=df , x='Age', hue='City_Category')
```

```python
sns.countplot(data=df , x='City_Category', hue='Gender')
```

```python
df.groupby('City_Category').sum()['Purchase'].plot(kind='pie', autopct='%.2f')
```

```python
df.groupby('City_Category').mean(numeric_only=True)['Purchase'].plot(kind='pie', autopct='%.2f')
```
